# pRF demo

Demo adapted from the Gardner lab pRF tutorial (http://gru.stanford.edu/doku.php/tutorials/pRF) by Justin Gardner and Dan Birman

```python
import numpy as np
import scipy
import scipy.io as sio
import matplotlib.pyplot as plt
import pylab
import time
import random
```

```python
# Load the data
pRFdata = sio.loadmat('pRF.mat')
```

```python
# Extract the data we need
stimImage = pRFdata['stimImage'] #stimulus
stimX = pRFdata['stimX'] #x coordinates of every point of the stimulus image
stimY = pRFdata['stimY'] #y coordinates of every point of the stimulus image
canonical = np.transpose(pRFdata['canonical']) #example hrf
canonical = canonical[:,0]

# two actual time series - one voxel in each
tSeries1 = pRFdata['tSeries1']; 
tSeries2 = pRFdata['tSeries2'];
tSeries1 = tSeries1[:,0]; 
tSeries2 = tSeries2[:,0];
tSeries = pRFdata['tSeries']; #time series for all voxels
```

```python
# Visualize stimulus to get a sense
fig = plt.figure
plt.ion()
for iFrame in range(30,32):
    plt.imshow(stimImage[:,:,iFrame],cmap="gray");
    plt.pause(.1)

plt.ioff()
```

## Compute Gaussian receptive field

```python
# Model the receptive field as a 2D gaussian
def makeGaussian(x,y,xCenter,yCenter,rfstd):
    return np.exp(-(np.power(x-xCenter,2) + np.power(y-yCenter,2))/(2*rfstd**2))
```

```python
# Let's start with a guess of what the receptive field is

# set variables for the center and width
xCenter = -12
yCenter = -3
rfstd = 3 #rf size

# compute the receptive field
rf = makeGaussian(stimX,stimY,xCenter,yCenter,rfstd)

# plot
plt.imshow(rf,cmap="hot");
```

## Compute the model neural response

```python
# Compute expected neural response given estimated RF and stimulus
# expected neural response = overlap between RF and stimulus
# (point by point multiplication of the stimulus image with the receptive fields and sum across all points)

modelNeuralResponse = np.zeros(stimImage.shape[2])
for iFrame in range(stimImage.shape[2]): #loop over frames in stim
  # multiply and sum
  modelNeuralResponse[iFrame] = np.sum(np.multiply(stimImage[:,:,iFrame],rf));

# plot what we got
plt.clf()
plt.plot(modelNeuralResponse, 'k-');
plt.title('Expected neural response')
```

## Compute the model BOLD response

```python
# Now we need to convolve this with the hemodynamic response function to get the predicted BOLD
# do the convolution
modelBoldResponse = np.convolve(modelNeuralResponse,canonical);
# plot
plt.clf()
plt.plot(modelBoldResponse, 'r-');
plt.title('Expected BOLD response')
```

```python
plt.clf();
plt.plot(modelNeuralResponse,'k-');
plt.plot(modelBoldResponse,'r-');
plt.title('Expected neural and BOLD responses')
plt.legend(['Neural','BOLD'],loc='upper right');
```

```python
# trim BOLD response to match neural response and transpose
modelBoldResponse = np.transpose(modelBoldResponse[0:modelNeuralResponse.shape[0]]);
```

## Compare the model to measured responses

```python
print(np.corrcoef(tSeries1,modelBoldResponse)[0,1])
print(np.corrcoef(tSeries2,modelBoldResponse)[0,1])
```

```python
# Let's plot the model and the time series together

# normalize the time series and the model so that the magnitudes match
# normalize time series
tSeries1 = (tSeries1-np.mean(tSeries1))\
            /np.sqrt(np.sum(np.power((tSeries1-np.mean(tSeries1)),2)));
tSeries2 = (tSeries2-np.mean(tSeries2))\
            /np.sqrt(np.sum(np.power((tSeries2-np.mean(tSeries2)),2)));
# and the model
modelBoldResponse = (modelBoldResponse-np.mean(modelBoldResponse))\
                    /np.sqrt(np.sum(np.power((modelBoldResponse-np.mean(modelBoldResponse)),2)));
# now plot
plt.clf()
plt.plot(tSeries1,'k-')
plt.plot(tSeries2,'b-')
plt.plot(modelBoldResponse,'r-')
plt.xlabel('Volume number')
plt.ylabel('Model and time series response')
plt.legend(['tSeries1','tSeries2','Model BOLD Response'],bbox_to_anchor=(1.05, 1))
```

## Finding the best model for a time series

Okay so we perfectly 'guessed' the receptive field for time series 1 but let's try and actually search for the best parameters for time series 2 using a simple grid search

```python
# Let's try x and y from -15 to 15 degrees in steps of 1 and standard deviations from 1 to 5 in steps of 1

k = 31*31*5 #k possible RFs
n = stimImage.shape[2] #length of model responses
modelResponses = np.zeros([k,n])
params = np.zeros([k,3])

modelNum = 1
for x in range(-15,15):
    for y in range(-15,15):
        for s in range(1,5):
            # compute the receptive field
            rf = makeGaussian(stimX,stimY,x,y,s)
            
            modelNeuralResponse = np.zeros(stimImage.shape[2])
            
            for iFrame in range(0,stimImage.shape[2]): #loop over stimulus
                # compute overlap and sum
                modelNeuralResponse[iFrame] = np.sum(np.multiply(stimImage[:,:,iFrame],rf))
                
            # now convolve with hrf
            modelBoldResponse = np.convolve(modelNeuralResponse,canonical)
            modelBoldResponse = np.transpose(modelBoldResponse[0:modelNeuralResponse.shape[0]]) #and trim
            
            # normalize
            modelBoldResponse = (modelBoldResponse-np.mean(modelBoldResponse))\
                    /np.sqrt(np.sum(np.power((modelBoldResponse-np.mean(modelBoldResponse)),2)));
            
            # save model and params
            modelResponses[modelNum,:] = modelBoldResponse
            params[modelNum,:] = [x,y,s]
            
            modelNum = modelNum + 1
            
    # display something so that we know that things are progressing
    print('Computed receptive fields for x =',x)
```

```python
# Compute correlation with every model response to find best model params for tSeries2
r = np.matmul(modelResponses,tSeries2); #can just use matrix product cuz we normalized
# which correlation is highest?
print(np.max(r))
bestModel = np.argmax(r)
# check what those parameters are
print(params[bestModel,:])
# plot the model and the response
plt.clf()
plt.plot(modelResponses[bestModel,:],'r-');
plt.plot(tSeries2,'k-');
plt.xlabel('Time in volumes');
plt.ylabel('Response');
plt.legend(['Best model','tSeries2']);
```

## Fit a bunch of voxels

```python
# Do the grid search to find the model with the best correlation with each time series
r = np.matmul(modelResponses,np.transpose(tSeries)) #we can use our previously computed model responses
maxR = np.max(r,0)
bestModel = np.argmax(r,0)
```

```python
# We can then plot the pRF centers and sizes for all the voxels
# All voxels are from right visual cortex so all pRFs should be in the left visual field

plt.clf()
for vox in range(0,bestModel.size): #cycle over voxels 
    if maxR[vox] > 0.5: #threshold so only voxels with correlations > .5 included
        # get the parameters
        centerX = params[bestModel[vox],0];
        centerY = params[bestModel[vox],1];
        rfstd = params[bestModel[vox],2];
        
        # now plot the RF
        x = np.linspace(-rfstd,rfstd,100)
        # plot the positive y-values
        plt.plot(x+centerX,np.sqrt(np.power(rfstd,2) - np.power(x,2))+centerY,'k-',linewidth=0.4);
        # plot the negative y-values
        plt.plot(x+centerX,-np.sqrt(np.power(rfstd,2) - np.power(x,2))+centerY,'k-',linewidth=0.4);
        # plot centers
        plt.plot(centerX,centerY,'ro', markersize=3)

plt.xlim(-20,20)
plt.ylim(-20,20)
plt.xlabel('Horizontal position (deg)');
plt.ylabel('vertical position (deg)');
```

```python
# Last sanity check: plot time series against model time course
# for voxel with low correlation to check that those are actually bad fits

# find all the voxels with correlations less than 0.1
print(np.where(maxR<0.1))
# let's just choose an arbitrary one to look at
checkVoxel = 240;
# find out what it's parameters were
params[bestModel[checkVoxel],:]
# plot model and time series
plt.clf()
plt.plot(tSeries[checkVoxel,:],'k-');
plt.plot(modelResponses[bestModel[checkVoxel],:],'r-');
# label graph
plt.xlabel('Time in volumes');
plt.ylabel('Response');
```
